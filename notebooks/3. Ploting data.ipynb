```python
import matplotlib as mpl
mpl.__version__
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
import numpy as np
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x, y)
```

```python
plt.plot(y) # same as above but without specifying x!
```

```python
plt.plot(y.cumsum())
```

```python
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal');
```

```python
cvalues = y.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(cvalues, 'b', lw=1.5) #lw = linewidth
plt.plot(cvalues, 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot');
```

## 2D-data sets

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot');
```

Use some annotations to distinguish between lines. Use the legend function to name the lines.

```python
def plot_2d_with_legend():
    plt.figure(figsize=(10, 6))
    plt.plot(y[:, 0], lw=1.5, label='1st')
    plt.plot(y[:, 1], lw=1.5, label='2nd')
    plt.plot(y, 'ro')
    plt.legend(loc=0)# loc=0 finds the best location in the graph
    plt.xlabel('index')
    plt.ylabel('value')
    plt.title('A simple plot');
```

```python
plot_2d_with_legend()
```

How data with different scaling can make some information to get lost.

```python
y[:, 0] = y[:, 0] * 100
plot_2d_with_legend()
```

## How to solve the issue with lost of information
* Use two y-axes
* Use two subplots

```python
fig, ax1 = plt.subplots() # new line in the code
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('values 1st')
plt.title('A simple plot')
ax2 = ax1.twinx() # new line in the code
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd');
```

Separate subplots

```python
plt.figure(figsize=(10, 6))
plt.subplot(211)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A simple plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value');
```

## Combining a line chart and a bar chart

```python
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('value')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd Data Set')
```

## Scatter plot

```python
y = np.random.standard_normal((1000, 2))
print('shape of y:', y.shape)
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plot');
```

## Histograms

```python
plt.figure(figsize=(10, 6))
plt.hist(y, label=['1st', '2nd'], bins=25)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram');
```

## Stacking histograms

```python
plt.figure(figsize=(10, 6))
plt.hist(y, label=['1st', '2nd'], color=['b', 'g'],
stacked=True, bins=20, alpha=0.5)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram');
```

## Box plot

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot(y)
plt.setp(ax, xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot');
```
